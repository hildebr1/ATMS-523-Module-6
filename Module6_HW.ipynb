```python
# Use the Storm Prediction Center tornado dataset https://www.spc.noaa.gov/wcm/data/1950-2021_actual_tornadoes.csv

# Also, use the ENSO, PDO, NAO, and AO dataframe from earlier module.
```

```python
import pandas as pd
```

```python
tornadoes = pd.read_csv('https://www.spc.noaa.gov/wcm/data/1950-2021_actual_tornadoes.csv')
print(tornadoes)
```

```python
# Open ENSO, PDO, NAO, AO data from Module 4 Notebook 1

import datetime
```

```python
# Read in ENSO, PDO, NAO, and AO data

enso = pd.read_csv('https://www.atmos.illinois.edu/~snesbitt/soi.dat',sep=r'\s+',header=None,skiprows=4,skipfooter=0, engine='python')
pdo = pd.read_csv('https://www.ncei.noaa.gov/pub/data/cmb/ersst/v5/index/ersst.v5.pdo.dat',sep=r'\s+',header=None,skiprows=2, engine='python')
nao = pd.read_csv('https://www.cpc.ncep.noaa.gov/products/precip/CWlink/pna/norm.nao.monthly.b5001.current.ascii.table',sep=r'\s+',header=None,skiprows=1, engine='python')
ao = pd.read_csv('https://www.cpc.ncep.noaa.gov/products/precip/CWlink/daily_ao_index/monthly.ao.index.b50.current.ascii.table',sep=r'\s+',header=None,skiprows=1, engine='python')
```

```python
# Get these into a usable format by creating a new data frame and reformatting the 12 x nyears array to a 1 x n observations array corresponding with each date.
# Starting with enso... 

enso_new=pd.DataFrame()
```

```python
enso_new
```

```python
enso[0].iloc[-1]
```

```python
enso_new['Date']=pd.date_range(start=datetime.datetime(enso[0].iloc[0],1,1),end=datetime.datetime(enso[0].iloc[-1],12,1),freq="MS")

enso_new = enso_new.set_index('Date')
```

```python
enso_new
```

```python
# Use stack to take the 2D array and create a 1D column

len(enso.loc[:,1:].stack().values)
```

```python
enso_new['ENSO']=enso.loc[:,1:].stack().values
```

```python
# Check data
enso_new
```

```python
# Repeat this for the other 3 indices

nao_new=pd.DataFrame()
pdo_new=pd.DataFrame()
ao_new=pd.DataFrame()

nao_new['Date'] = pd.date_range(start=datetime.datetime(nao[0].iloc[0],1,1),end=datetime.datetime(nao[0].iloc[-1],12,1),freq="MS")
pdo_new['Date'] = pd.date_range(start=datetime.datetime(pdo[0].iloc[0],1,1),end=datetime.datetime(pdo[0].iloc[-1],12,1),freq="MS")
ao_new['Date'] = pd.date_range(start=datetime.datetime(ao[0].iloc[0],1,1),end=datetime.datetime(ao[0].iloc[-1],12,1),freq="MS")

nao_new = nao_new.set_index('Date')
pdo_new = pdo_new.set_index('Date')
ao_new = ao_new.set_index('Date')

nao_new['NAO']=nao.loc[:,1:].stack(dropna=False).values
pdo_new['PDO']=pdo.loc[:,1:].stack(dropna=False).values
ao_new['AO']=ao.loc[:,1:].stack(dropna=False).values
```

```python
# Merge the 4 indices into 1 dataset

newdf_all = pd.merge(enso_new,pdo_new, left_index=True, right_index=True)
```

```python
newdf_all
```

```python
import numpy as np
newdf_all.loc[newdf_all['PDO'] > 90.,'PDO'] = np.nan
```

```python
newdf_all.loc[newdf_all['ENSO'] <= -9.9,'ENSO'] = np.nan
```

```python
newdf_all = pd.merge(newdf_all,nao_new, left_index=True, right_index=True)
newdf_all.loc[newdf_all['NAO'] <= -99.9,'NAO'] = np.nan
```

```python
newdf_all = pd.merge(newdf_all,ao_new, left_index=True, right_index=True)
newdf_all.loc[newdf_all['AO'] <= -99.9,'AO'] = np.nan
```

```python
# All 4 indices should now be in newdf_all, with NaNs appropriately filled in

newdf_all
```

```python
# 1 Aggregate the tornado data for Illinois into a monthly count of tornadoes
```

```python
print(tornadoes)
```

```python
# Filter the dataset for IL only
tornadoes_il = tornadoes[tornadoes["st"] == "IL"].copy()

# Make sure the date is in the correct datetime format
tornadoes_il["date"] = pd.to_datetime(tornadoes_il["date"])

# Set the timespan starting in January 1951 since that is when the indices data starts, and ending in August 2021 since the tornado dataset doesn't have a full month of data for September 2021.
tornadoes_il = tornadoes_il[(tornadoes_il["date"] >= "1951-01-01") & (tornadoes_il["date"] <= "2021-08-31")]

# Convert to month format YYYY-MM-01 to match the indices data
tornadoes_il["month"] = tornadoes_il["date"].dt.to_period("M").dt.to_timestamp()

# Aggregate the count of tornadoes per month
monthly_tornadoes = tornadoes_il.groupby("month").size().to_frame(name="IL_monthly_count")

print(monthly_tornadoes)
```

```python
# 2. Merge this monthly tornado dataset with the ENSO, PDO, NAO, and AO dataframe created in Module 4 Notebook 1.
```

```python
# Date is the index for both dataframes. 
# Use a left join so that index data is only joined where there is a monthly tornado count

alldata = monthly_tornadoes.join(newdf_all, how="left")

print(alldata)
```

```python
# 3. Create a random forest regression model and use features that include the 4 climate indices to predict the monthly number of tornadoes.  
# Create a scatter plot of the predicted and observed number of tornadoes over these states.  
# Report the correlation coefficient and RMSE of your model.
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
# Set the climate indices as the features and tornado count as the target
X = alldata[['ENSO', 'PDO', 'NAO', 'AO']]
y = alldata['IL_monthly_count']

# Use 80% for train, 20% for test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

print('Train size:', X_train.shape, 'Test size:', X_test.shape)
```

```python
# Fit the random forest regressor model

rf = RandomForestRegressor(n_estimators=500, max_depth=None, random_state=42)

rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
corr, _ = pearsonr(y_test, y_pred)
```

```python
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
print(f"Correlation: {corr:.3f}")
print(f"RMSE: {rmse:.3f}")
```

```python
# A Pearson's correlation coefficient of 0.029 indicates a weak correlation. (Close to 1 is strongest correlation)
# A root mean squared error of 7.598 is pretty high. (The lower, the better)
```

```python
# Scatter plot of predicted vs. observed tornadoes in IL: 

plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred)
plt.plot([0, max(y_test.max(), y_pred.max())],
         [0, max(y_test.max(), y_pred.max())],
         linestyle='--')
plt.xlabel("Observed")
plt.ylabel("Predicted")
plt.title("Observed vs Predicted IL Monthly Tornado Counts")
plt.grid(True)
plt.show()
```

```python
# 4. Create a one-hot encoding for the month of the year.  
# Repeat the scatterplot and compare the RMSE and correlation coefficient values.
```

```python
# Extract month
alldata["month"] = alldata.index.month
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
OHE = OneHotEncoder(sparse_output=False, dtype=int)

# Fit and transform the month column (needs 2D array)
month_encoded = OHE.fit_transform(alldata[["month"]])
```

```python
# Name columns for each month
month_cols = OHE.get_feature_names_out(["month"])

# Convert to DataFrame
month_df = pd.DataFrame(month_encoded, index=alldata.index, columns=month_cols)

# Combine this with the original data
alldata_encoded = pd.concat([alldata, month_df], axis=1)
```

```python
print(alldata_encoded)
```

```python
# Repeat random forest and scatterplot

# Set the climate indices and months as the features and tornado count as the target
X2 = alldata_encoded[['ENSO', 'PDO', 'NAO', 'AO', 'month_1', 'month_2', 'month_3', 'month_4', 'month_5', 'month_6', 'month_7', 'month_8', 'month_9', 'month_10', 'month_11', 'month_12']]
y2 = alldata_encoded['IL_monthly_count']

# Use 80% for train, 20% for test
X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2, test_size=0.2)

print('Train size:', X_train2.shape, 'Test size:', X_test2.shape)
```

```python
# Fit the random forest regressor model

rf_encoded = RandomForestRegressor(n_estimators=500, max_depth=None, random_state=42)

rf_encoded.fit(X_train2, y_train2)
```

```python
y_pred2 = rf_encoded.predict(X_test2)
```

```python
corr2, _ = pearsonr(y_test2, y_pred2)
```

```python
rmse2 = np.sqrt(mean_squared_error(y_test2, y_pred2))
```

```python
print(f"Correlation: {corr2:.3f}")
print(f"RMSE: {rmse2:.3f}")
```

```python
# Compare the RMSE and correlation coefficient values: 

print(f"Correlation with just random forest: {corr:.3f}")
print(f"RMSE with just random forest: {rmse:.3f}")

print(f"Correlation with one hot encoding: {corr2:.3f}")
print(f"RMSE with one hot encoding: {rmse2:.3f}")
```

```python
# The correlation coefficient improved slightly, but the RMSE increased. Adding in one hot encoding did not improve the model significantly. 
```

```python
# Scatter plot of predicted vs. observed tornadoes in IL when using one hot encoding: 

plt.figure(figsize=(8, 6))
plt.scatter(y_test2, y_pred2)
plt.plot([0, max(y_test2.max(), y_pred2.max())],
         [0, max(y_test2.max(), y_pred2.max())],
         linestyle='--')
plt.xlabel("Observed")
plt.ylabel("Predicted")
plt.title("Observed vs Predicted IL Monthly Tornado Counts")
plt.grid(True)
plt.show()
```

```python
# 5. Perform a feature importance and multipass permutation analysis of your features.
```

```python
# Feature importance:
featureimportances = pd.Series(rf.feature_importances_, index=X.columns).sort_values(ascending=True)

print("RF Feature Importances:")
print(featureimportances)
```

```python
# Multipass permutation analysis:
from sklearn.inspection import permutation_importance
```

```python
result = permutation_importance(
    rf, X_test, y_test, n_repeats=10, random_state=42, n_jobs=2
)

sorted_importances_idx = result.importances_mean.argsort()
importances = pd.DataFrame(
    result.importances[sorted_importances_idx].T,
    columns=X.columns[sorted_importances_idx],
)

print("Permutation Importances:")
print(importances)
```

```python
import shap
```

```python
# Create SHAP explainer
explainer = shap.TreeExplainer(rf)

# Compute SHAP values for test set
shap_values = explainer.shap_values(X_test)

# Plot
shap.summary_plot(shap_values, X_test, show=False)
```

```python
# 7. Create a SHAP Feature Dependence for the ENSO index.

shap.dependence_plot("ENSO", shap_values, X_test)
```
